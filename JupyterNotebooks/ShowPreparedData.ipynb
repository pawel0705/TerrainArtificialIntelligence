```python
import matplotlib.pyplot as plt
import random
import numpy as np
import h5py
```

```python
f = h5py.File('dataset2.hdf5', 'r')
src_images, tar_images = f['sketches'], f['heightmaps']

# plot source images
n_samples = 3

index = []
for i in range(n_samples):
    index.append(random.randint(0, 20))

for i in range(n_samples):
    plt.subplot(2, n_samples, 1 + i)
    plt.axis('off')
    plt.imshow(((src_images[index[i]] + 1.0) / 2.0))
    print(index[i])

# plot target image
for i in range(n_samples):
    plt.subplot(2, n_samples, 1 + n_samples + i)
    plt.axis('off')
    plt.imshow(((tar_images[index[i]] + 1.0) / 2.0))
    
plt.rcParams['figure.figsize'] = [40, 40]
plt.show()
```

```python
f.close()
```
